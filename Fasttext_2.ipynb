```python
#1.데이터 불러오기(데이터형확인)(dim,shape)
from google.colab import drive

# Google Drive 마운트
drive.mount('/content/drive')

# Google Drive에 있는 파일 읽기
file_path = '/content/drive/MyDrive/Colab Notebooks/db.sheet.csv'
dataset = pd.read_csv(file_path)

print(dataset.dtypes)# 각 열의 데이터 타입 확인
print(dataset.shape)# 데이터의 행과 열의 개수 확인

#titleSet keywordSet나누기
X = dataset.iloc[:, 1] #titleSet
y1 = dataset.iloc[:, 2] #keywordSet1
y2 = dataset.iloc[:, 3] #keywordSet2
y3 = dataset.iloc[:, 4] #keywordSet3

print("X_dataset")
print(X)
print("X_shape")
print(X.shape)

print("y1_dataset")
print(y1)
print("y1_shape")
print(y1.shape)

print("y2_dataset")
print(y2)
print("y3_dataset")
print(y3)
```

```python
# Kiwi 형태소 분석기 초기화
kiwi = Kiwi()

# 각 행의 키워드를 저장할 리스트
X_kiwi = []

# 각 행을 처리하여 키워드 추출 후 저장
for title in X:
    # Kiwi 형태소 분석 수행
    kiwi_result = kiwi.analyze(title)

    # 명사 추출 함수
    def noun_extractor_kiwi(kiwi_result):
        nouns = []
        for token, pos, _, _ in kiwi_result[0][0]:
            if len(token) != 1 and (pos.startswith('N') or pos.startswith('SL')):
                nouns.append(token)
        return nouns

    # 명사 추출
    nouns = noun_extractor_kiwi(kiwi_result)

    # 추출된 명사를 X_kiwi에 추가
    X_kiwi.append(nouns)

for i in range(len(X_kiwi)):
  print(X_kiwi[i])

print(len(X_kiwi))
#해당 X_kiwi값을 드라이브의 DB.XLSX의 KIWI파일로 업로드
```

```python
#KIWI=KIWI_COMBINED(목적; 코버트 사용을 위해 리스트에서 문자열로 JOIN
# x 리스트에 있는 문자열들을 join하여 하나의 문자열로 합치기

X_kiwi_combined = [' '.join(sentence) for sentence in X_kiwi]
print(len(X_kiwi_combined))
print(X_kiwi_combined)

#k_kiwi값 엑셀에 자동저장(num=217,header가 추가됨)
!pip install xlsxwriter
df_kiwi = pd.DataFrame(X_kiwi)
file_path = '/content/drive/MyDrive/Colab Notebooks/kiwi.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    df_kiwi.to_excel(writer, sheet_name='kiwi_auto', index=False)

print(len(X_kiwi_combined))
```

```python
#weight_set초기화
weight_set=[]

#weight_set에서 각 문자열을 가져와서 단어의 개수 확인(중복 제거 이전의 sentence length)
sentence_lengths = []
for sentence in X_kiwi_combined:
    words = sentence.split()
    sentence_length = len(words)
    sentence_lengths.append(sentence_length)

print("Total sentences:", len(X_kiwi_combined))
print("Sentence lengths:", sentence_lengths)
```

```python
from gensim.models import FastText
from gensim.models.word2vec import LineSentence

# 파일 경로 설정
file_path = '/content/drive/MyDrive/Colab Notebooks/db.sheet.csv'

# 단어 집합 구축 및 모델 학습
sentences = LineSentence(file_path)
model = FastText(vector_size=50, window=5, min_count=10)
model.build_vocab(sentences)
model.train(sentences, total_examples=model.corpus_count, epochs=10)
print(model)
```

```python
from gensim.models import FastText
import pandas as pd
import csv

# CSV 파일 경로
file_path = '/content/drive/MyDrive/Colab Notebooks/kiwi_auto.csv'

# CSV 파일을 읽어와 텍스트로 변환
text_corpus = []
with open(file_path, 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        # CSV 파일의 각 행을 공백으로 구분된 하나의 텍스트로 결합
        text = ' '.join(row)
        text_corpus.append(text)

# FastText 모델 학습
model = FastText(sentences=text_corpus, vector_size=100, window=5, min_count=5, workers=4, sg=1)
# sentences 매개변수에는 텍스트 리스트가 들어가야 합니다.

# 모델 확인
print(model)


##리스트로 변환해와야 한다고?
#Text8Corpus가 텍스트를 처리하는 방식과 관련이 있을 것으로 보입니다. Text8Corpus는 각 문장이 단어로 구성된 리스트여야 하지만, 문자열로 구성되어 있어서 발생하는 문제로 추정됩니다.
#문제를 해결하려면 Text8Corpus가 올바른 형식의 데이터를 입력받을 수 있도록 데이터를 전처리하여 리스트로 구성된 형태로 제공해야 합니다.
#그 후에 모델을 학습시키면 이러한 경고가 해결될 것입니다.
```

```python
from gensim.models import FastText

# 모델을 저장하고 로드하는 코드
model.save('fasttext_model')
saved_model = FastText.load('fasttext_model')

# '연구원' 단어의 벡터를 가져오기
word_vector = saved_model.wv['영어']
print(word_vector)
```

```python
similar_words = saved_model.wv.most_similar("영어", topn=15)
for word, similarity in similar_words:
    print(f"{word}: {similarity}")
```

```python
similarity1 = saved_model.wv.similarity("인재", "프로그램")
similarity2 = saved_model.wv.similarity("인재", "상담")

print(f"Similarity between '인재' and '프로그램': {similarity1}")
print(f"Similarity between '인재' and '상담': {similarity2}")
```
