# 2.3 konlpy를 사용하여 목적어에 가중치 부여


*    복합명사를 분해함
*   목적어만 식별 불가 -> 명사 중 키워드 선택 어려움


```python
!pip install konlpy
```

```python
from konlpy.tag import Komoran

def assign_weight_to_noun(sentence, noun_weight=2.0):
    # KoNLPy 형태소 분석기 초기화
    komoran = Komoran()

    # 형태소 분석
    morphs = komoran.pos(sentence)

    # 명사 리스트 생성
    nouns = [word for word, pos in morphs if pos.startswith('N')]

    # 목적어에 대한 가중치 부여
    weighted_morphs = [(word, noun_weight) if word in nouns else (word, 1.0) for word, _ in morphs]

    # 가중치를 반영한 목적어를 문자열로 변환
    result_sentence = ' '.join([f"{word}({weight})" for word, weight in weighted_morphs])

    return result_sentence

# 예시 문장
sentence = "자연어 처리는 인공 지능의 중요한 부분입니다."

# 목적어에 가중치 부여
result = assign_weight_to_noun(sentence)

print(result)
```
